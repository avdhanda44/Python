# Seaborn

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn builds on top of Matplotlib and introduces additional features and capabilities to make creating plots and visualizations easier and more intuitive.

Seaborn provides several types of plots, including:

- Scatter plots
- Line plots
- Histograms
- Bar plots
- Heatmaps
- Pair plots
- Violin plots
- Box plots
- Joint plots
- Facet grids

Some of the key features of Seaborn include:

- Built-in themes for quickly customizing the look of plots
- Color palettes for adding color to plots
- Support for creating complex multi-plot grids and subplots
- Tools for visualizing and exploring distributions of data
- Support for plotting statistical models and visualizing relationships between variables


```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

## Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joint Plot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair Plot

A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df,hue='sex')
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- countplot
- bar plot
- boxplot
- violinplot

### Count plot
A count plot is a way to visualize the distribution of categorical data. It displays the number of observations in each category.

```python
sns.countplot('sex',data=df)
```

## Bar plot
 A bar plot is a way to visualize the relationship between a numerical and a categorical variable. It displays the mean (or another summary statistic) of the numerical variable in each category.

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

### Box Plot
A box plot is a way to visualize the distribution of numerical data across categories. It displays the median, quartiles, and outliers of the data in each category.

```python
 sns.boxplot('smoker','total_bill', data=df)
```

### Violin Plot
A violin plot is similar to a box plot, but it also shows the kernel density estimate of the underlying distribution. This can provide more information about the shape of the distribution.

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
